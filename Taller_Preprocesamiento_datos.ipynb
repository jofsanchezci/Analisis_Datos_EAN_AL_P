**Punto 1**

Generar dos arreglos aleatorios de 50 datos:



*   Semilla 0
*   Distribución Normal
*   Media 50
*   Desviación tipica 5
*  Realizar: Diagrama de dispersión
*  Histograma de frecuencia para cada arreglo
*   Diagrama de caja para cada arreglo




```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
arr1=np.random.normal(50,5,50)
arr2=np.random.normal(50,5,50)
```

```python
print(arr1)
print('---------------------------------')
print(arr2)
```

```python
print(arr1.shape)
print(arr1.ndim)
print(np.mean(arr1))
print(np.mean(arr2))
```

```python
#Grafico de dipersión
plt.scatter(arr1,arr2,c='red',marker='^',alpha=0.5)
plt.xlabel('Arreglo 1')
plt.ylabel('Arreglo 2')
plt.title('Grafica de dispersión')
plt.grid()
plt.show()
```

```python
plt.plot(arr1,arr2,'b--')
plt.xlabel('Arreglo 1')
plt.ylabel('Arreglo 2')
plt.title('Grafica de lineas')
plt.grid()
plt.show()
```

```python
#Histograma de frecuencias
plt.hist(arr1,bins=10)
plt.xlabel('Intervalos')
plt.ylabel('Frecuencias')
plt.title('Histograma de frecuencias Arreglo 1')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].hist(arr1,bins=10)
ax[0].set_title('Histograma Arr 1', fontsize=10)
ax[0].grid()
ax[1].hist(arr2,bins=10)
ax[1].set_title('Histograma Arr 2', fontsize=10)
ax[1].grid()

plt.show()
```

```python
print(ax)
print(type(ax))
print(ax.shape)
print(ax.ndim)
```

```python
#Diagrama de Caja
fig, ax = plt.subplots(1, 2)
ax[0].boxplot(arr1)
ax[0].set_title('Box Plot Arr 1', fontsize=10)
ax[0].grid()
ax[1].boxplot(arr2)
ax[1].set_title('Boxplot Arr 2', fontsize=10)
ax[1].grid()
plt.show()
```

**Punto 2**

*Datos Anexos*

* Diagrama de dispersion entre los puntajes
* Diagrama de caja para los puntajes
* Grafica de líneas entre cada puntaje y cada Id



```python
#Cargar el data Set
import pandas as pd
# Carga el archivo de Excel en un DataFrame de Pandas
df = pd.read_excel('data_set_A9.xlsx',sheet_name='Hoja1')
# comvierte los datos de una columuna de un dataFrame a una lista
idl=list(df['codigo_dane'])
punl=list(df['muestra'])
#Pasamos las listas a arreglos de numpy
nid=np.array(idl)
npun=np.array(punl)

print(nid.shape)
print(npun.shape)
```

```python
#Diagrama de dispersión
import matplotlib.pyplot as plt
plt.scatter(nid,npun)
plt.xlabel('Codigo Dane')
plt.ylabel('Puntaje')
plt.title('Diagrama de dispersión')
plt.grid()
plt.show()
```

```python
#Grafica de lineas
plt.plot(nid,npun, ':r')
plt.xlabel('Codigo Dane')
plt.ylabel('Puntaje')
plt.title('Diagrama de lineas')
plt.grid()
plt.show()
```

```python
#Comparativa entre un diaagrama de dispersión y de lineas
plt.plot(nid,npun, ':r')
plt.xlabel('Codigo Dane')
plt.scatter(nid,npun,c='g',marker='^',alpha=0.3)
plt.ylabel('Puntaje')
plt.title('Diagrama de comparación')
plt.grid()
plt.show()
```

```python
#Histograma de frecuencias para los puntaje
plt.hist(npun,bins=10)
plt.title('Histograma de frecuencias para Puntajes')
plt.xlabel('Intrevalos')
plt.ylabel('Frecuencias')
plt.grid()
plt.show()
```

```python
#Histograma de caja para los puntaje
plt.boxplot(npun)
plt.title('Diagrama de Caja para Puntajes')
plt.grid()
plt.show()
print('La media de los puntajes: ', np.mean(npun))
```

```python
#Comparación entre Histograma y diagrama de caja
fig, ax = plt.subplots(1, 2)
ax[0].hist(npun, bins=20)
ax[0].axvline(x=np.mean(npun), ymin=0.0, ymax=0.9,color='r')
ax[0].axvline(x=np.quantile(npun,0.75), ymin=0.0, ymax=0.9,color='c')
ax[0].axvline(x=np.quantile(npun,0.25), ymin=0.0, ymax=0.9,color='y')

ax[0].set_title('Histograma', fontsize=10)
ax[0].grid()
ax[1].boxplot(npun)
ax[1].set_title('Boxplot', fontsize=10)
ax[1].grid()
plt.show()
```
